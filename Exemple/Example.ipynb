```python
import pyCraFT.runcraft as RC
```

```python
adr_C_output='data/'
time_output='09.20000000e+04'
```

```python
data=RC.runcraft(adr_C_output,time_output)
```

```python
basal_act=data.gamma.gamma_activity()
```

```python
basal_act.plot()
```

```python
data.gamma.vpstrain(data).eqVonMises().plot()
```

```python
data.strain.eqVonMises().plot()
```

```python
import matplotlib.pyplot as plt
plt.plot(data.strain.eqVonMises().field.flatten(),data.gamma.vpstrain(data).eqVonMises().field.flatten(),'+')
plt.plot([0,0.2],[0,0.2],'r')
```
